```python
import matplotlib.pyplot as plt
from reservoirpy.datasets import rabinovich_fabrikant, santafe_laser, rossler

data = rabinovich_fabrikant(5000, h=0.1, gamma=0.1, alpha=0.98, x0=[0.1, -0.1, 0.1])
time = range(len(data))
data.shape
```

```python
for i in range(data.shape[1]):
   plt.plot(time, data[:, i], label=f"Variable {i+1}")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Rabinovich-Fabrikant System")
plt.legend()
plt.show()
```

```python
data = santafe_laser()
time = range(len(data))
plt.plot(time[:1000], data[:1000])
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Santafe Laser System")
plt.show()
```

```python
data = rossler(25000)
data = data[20000:21000]
time = range(len(data))

for i in range(data.shape[1]):
   plt.plot(time, data[:, i], label=f"Variable {i+1}")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Rabinovich-Fabrikant System")
plt.legend()
plt.show()
```
